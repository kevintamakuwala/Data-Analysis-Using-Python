```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split as tts
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
data,target = iris.data, iris.target
xTrain, xTest, yTrain, yTest = tts(data,target,test_size = .4, random_state = 42)

model = KNeighborsClassifier(n_neighbors = 3)
model.fit(xTrain,yTrain)
ans = model.predict(xTest)

cm=confusion_matrix(ans,yTest)
print(accuracy_score(ans,yTest))
print(cm)

plt.scatter(xTest[:,0],yTest,color='red')
plt.scatter(xTest[:,0],ans,color='green')
plt.plot(xTest[:,0],ans,color='blue')
plt.show()
```
