```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
diabetes = load_diabetes()
data=diabetes.data
target=diabetes.target
df=pd.DataFrame(data,columns=diabetes.feature_names)
# print(df.keys())
# print(diabetes.feature_names)
# print(diabetes.target)

db = data[:np.newaxis,2]
db_X_train=db[:-30].reshape(-1,1)
db_X_test=db[-20:].reshape(-1,1)
db_Y_train =target[:-30]
db_Y_test = target[-20:]

linearRegression = LinearRegression()
linearRegression.fit(db_X_train,db_Y_train)
predictions=linearRegression.predict(db_X_test)

plt.scatter(db_X_test,db_Y_test,s=5)
plt.plot(db_X_test,predictions,"r--")
plt.show()
print(linearRegression.score(db_X_test,db_Y_test))
```
